#### Scrapping the details for flight price prediction

```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
link = "https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=23%2F10%2F2021&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-home&unqvaldesktop=265183604656"
driver.get(link)
```

```python
name = []                  
date = []                 
source = []               
destination = []                      
dep_time = []                 
arr_time = []                 
duration = []         
stop = []       
price = []  

#Scrapping the details 23rd October

try:
    n = driver.find_elements_by_xpath("//div[@class = 'fs-13 airline-name no-pad col-8']//span[@class = 'i-b text ellipsis']")

    for i in n:
        name.append(i.text)

except NoSuchElementException:
    name.append('NaN')
    
    
    
time.sleep(2)
try:
    d = driver.find_element_by_xpath("//ul[@class = 'full-width no-wrap ovf-hidden']//li[2]//div//p[1]")
    linked =  driver.find_elements_by_xpath("//div[@class = 'fs-13 airline-name no-pad col-8']//span[@class = 'i-b text ellipsis']")
    for j in range(0,len(linked)):
        date.append(d.text)
except NoSuchElementException:
    date.append('NaN')
    
    
time.sleep(2)
try:
    de = driver.find_elements_by_xpath("//div[@class = 'i-b col-4 no-wrap text-right dtime col-3']//div//div[@class = 'i-b pr']")

    for i in de:
        dep_time.append(i.text)
except NoSuchElementException:
    dep_time.append('NaN')
    

time.sleep(2)
try:
    a = driver.find_elements_by_xpath("//div[@class = 'i-b pdd-0 text-left atime col-5']//p[1]")
    for i in a:
        arr_time.append(i.text)
except NoSuchElementException:
    arr_time.append('NaN')
    
time.sleep(2)
try:
    s = driver.find_elements_by_xpath("//div[@class = 'i-b col-4 no-wrap text-right dtime col-3']//p[@class = 'fs-10 font-lightgrey no-wrap city ellipsis']")

    for i in s:
        source.append(i.text)
except NoSuchElementException:
    source.append('NaN')
    
    
    
time.sleep(2)
try:
    d = driver.find_elements_by_xpath("//div[@class = 'i-b pdd-0 text-left atime col-5']//p[@class = 'fs-10 font-lightgrey no-wrap city ellipsis']")

    for i in d:
        destination.append(i.text)
except NoSuchElementException:
    destination.append('NaN')
    
    
time.sleep(2)
try:
    du = driver.find_elements_by_xpath("//div[@class = 'stop-cont pl-13']//p[@class = 'fs-12 bold du mb-2']")

    for i in du:
        duration.append(i.text)
except NoSuchElementException:
    duration.append('NaN')  
    

    
    
time.sleep(2)
try:
    st = driver.find_elements_by_xpath("//div[@class = 'stop-cont pl-13']//div[@class = ' font-lightgrey fs-10 tipsy i-b fs-10']//span")

    for i in st:
        stop.append(i.text)
except NoSuchElementException:
    stop.append('NaN')
    
    
    
    
time.sleep(2)
try:
    fares = driver.find_elements_by_xpath("//div[@class = 'i-b tipsy fare-summary-tooltip fs-18']")

    for i in fares:
        price.append(i.text)
except NoSuchElementException:
    price.append('NaN')
```

```python
print(len(name),len(date),len(source),len(destination),len(dep_time),len(arr_time),len(duration),len(stop),len(price))
```
